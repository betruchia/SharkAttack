```python
clean_attack = attacks.groupby("Country").agg({"Fatal (Y/N)":"count"})
clean_attack.reset_index(inplace=True)
clean_attack.rename(columns = {"Country":"country", "Fatal (Y/N)":"attacks"}, inplace = True)
```

```python
attacks.loc[attacks["Fatal (Y/N)"]=="Y", "deaths"] = 1
attacks.loc[attacks["Fatal (Y/N)"]=="N", "deaths"] = 0
```

```python
clean_fatal = attacks.groupby("Country").agg({"deaths":"sum"})
clean_fatal.reset_index(inplace=True)
clean_fatal.rename(columns = {"Country":"country"}, inplace = True)
clean_fatal["deaths"] = clean_fatal["deaths"].astype(int)
```

```python
shark_clean = pd.merge(left=clean_attack, right=clean_fatal) ####################################
```

```python
shelf = pd.read_csv("EEZ.csv")
```

```python
shelf.head()
```

```python
shelf["Country"] = shelf["Country"].str.upper()
```

```python
shelf["Country"] = shelf["Country"].str.replace("UNITED STATES", "USA")
```

```python
shelf.drop(columns = ["Rank", "EEZ km2", "EEZ+TIA km2"], inplace = True)
```

```python
shelf.loc[shelf["Country"]=="NEW ZEALAND", "Shelf km2"] = 274395
shelf.drop(shelf.loc[shelf["Country"] == "NIUE", "Country"].index, inplace = True)
shelf.drop(shelf.loc[shelf["Country"] == "COOK ISLANDS", "Country"].index, inplace = True)
```

```python
top_shelves = shelf.sort_values(by = "Shelf km2", ascending = False).head(20)
```

```python
top_shelves.reset_index(drop=True,inplace=True)
```

```python
top_shelves.reset_index(inplace=True)
```

```python
top_shelves.rename(columns = {"index":"area_rank", "Country":"country", "Shelf km2":"shelf_area(km²)"}, inplace=True)
```

```python
top_shelves["area_rank"] = top_shelves["area_rank"].apply(lambda x : x+1)
```

```python
proportional_shark = pd.merge(left=shark_clean,right=top_shelves, how="inner")
```

```python
proportional_shark["proportional_attacks"]=proportional_shark["attacks"]/proportional_shark["shelf_area(km²)"]
```

```python
proportional_shark["proportional_deaths"]=proportional_shark["deaths"]/proportional_shark["shelf_area(km²)"]
```

```python
proportional_shark["fatality_rate"]=proportional_shark["deaths"]/proportional_shark["attacks"]
```

```python
proportional_shark.sort_values("area_rank",inplace=True)
```

```python
proportional_shark.set_index("area_rank", inplace=True)
```

```python
proportional_shark = proportional_shark[["country","shelf_area(km²)","attacks","deaths","proportional_attacks","proportional_deaths","fatality_rate"]]
```

```python
proportional_shark.sort_values("proportional_attacks", ascending=False)
```

```python
proportional_shark.sort_values("proportional_deaths", ascending=False)
```

```python
proportional_shark.sort_values("fatality_rate", ascending=False)
```
